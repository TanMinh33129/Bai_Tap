### 3. Tính giá trị trung bình của cột <AGE>.

### 13 Thay thế các giá trị của cột <Age> lớn hơn 40 bằng giá trị trung bình của cột đó

### 14.  Biết công thức tính chỉ số BMI như sau: 
### BMI = Weight(kg)/Height2 (m2)  
### Tạo một cột mới trong Dataframe với tên <cs_BMI>, tính chỉ số BIM mỗi người

```python
# Tính chỉ số BMI và thêm vào DataFrame
df["cs_BMI"] = df["Weight"] / (df["Height"] **2 )

# Hiển thị DataFrame sau khi thêm cột cs_BMI
print("DataFrame sau khi thêm cột chỉ số BMI:")
print(df)
```

### 15 Theo thang đo kết quả tính BMI của châu Á: 
### Chỉ số BMI dưới 18,5 là thiếu cân, 18.5-22,9 là bình thường, 23-24,9 là thừa cân, từ 25 trở lên là béo phì.  
### Tạo cột với tên <Phan_Loai> với các mức ‘0’,’1’,’2’,’3’ tương ứng với mức chỉ  số BMI trên.

```python
# Hàm phân loại BMI theo mức 0, 1, 2, 3
def classify_bmi(bmi):
    if bmi < 18.5:
        return "0"  # Thiếu cân
    elif 18.5 <= bmi < 23:
        return "1"  # Bình thường
    elif 23 <= bmi < 25:
        return "2"  # Thừa cân
    else:
        return "3"  # Béo phì

# Áp dụng hàm để tạo cột 'Phan_Loai'
df["Phan_Loai"] = df["cs_BMI"].apply(classify_bmi)

# Hiển thị DataFrame sau khi thêm cột 'Phan_Loai'
print("DataFrame sau khi phân loại chỉ số BMI:")
print(df)
```

### 16  Chuẩn hóa giá trị của cột <cs_BMI> chuẩn hóa giá trị trong đoạn từ 0 tới 1. 

```python

# Tạo cột 'cs_BMI' nếu chưa có
if "cs_BMI" not in df.columns:
    df["cs_BMI"] = df["Weight"] / (df["Height"] ** 2)

# Loại bỏ khoảng trắng trong tên cột (tránh lỗi do dữ liệu nhập vào)
df.columns = df.columns.str.strip()

# Chuẩn hóa cột 'cs_BMI' về khoảng [0,1]
df["cs_BMI_norm"] = (df["cs_BMI"] - df["cs_BMI"].min()) / (df["cs_BMI"].max() - df["cs_BMI"].min())

# Hiển thị DataFrame 
print("DataFrame sau khi chuẩn hóa cs_BMI:")
print(df)

```

### 17  Cho biết: 
### + ‘Mr‘tương ứng với ‘male’, 
### + ’Mrs’ và ‘Miss’ tương ứng với ‘female’.  
### Tạo cột với tên <GT>, Sau khi thực hiện, xóa các từ Mr, Mrs, Miss nếu trong  cột <Name> còn các từ đó.

```python
import pandas as pd

# Tạo DataFrame 
data = {
    "Name": ["Mr. Harris", "Mrs. Bradley", "Miss. Laina", "Mrs. Heath", "Mr. William",
             "Mr. James", "Mr. Timothy", "Mrs. Oscar", "Mrs. Nicholas", "Miss. Marguerite",
             "Mr. William", "Miss. Hulda", "Mr. Lawrence", "Mr. Wilam"],
    "Height": [1.80, 1.61, 1.78, 1.74, 1.55, 1.9, 1.61, 1.40, 1.92, 1.77, 1.55, 1.85, 1.64, 1.77],
    "Weight": [90, 87, 64, 70, 100, 106, 82, 55, 77, 60, 100, 109, 80, 66],
    "Age": [32, 45, 22, 52, 48, 21, 28, 69, 58, 21, 48, 33, 25, 70]
}

df = pd.DataFrame(data)

import re

# Hàm xác định giới tính
def get_gender(name):
    if name.startswith("Mr"):
        return "male"
    elif name.startswith("Mrs") or name.startswith("Miss"):
        return "female"
    return "unknown"

# Tạo cột 'GT' xác định giới tính
df["GT"] = df["Name"].apply(get_gender)

# Loại bỏ tiền tố 'Mr', 'Mrs', 'Miss' khỏi cột 'Name'
df["Name"] = df["Name"].apply(lambda x: re.sub(r"^(Mr|Mrs|Miss)\.?\s*", "", x))

# Hiển thị DataFrame sau khi cập nhật
print("DataFrame sau khi thêm cột giới tính và loại bỏ tiền tố:")
print(df)

```

### 18 Vẽ biểu đồ tròn Cho GT 

```python
import matplotlib.pyplot as plt

gt_counts = df["GT"].value_counts()

# Vẽ biểu đồ tròn
plt.figure(figsize=(6, 6))
plt.pie(gt_counts, labels=gt_counts.index, autopct="%1.1f%%", colors=["blue", "pink"])
plt.title("Tỷ lệ giới tính (GT)")
plt.show()
```

### 19 Trực quan hóa dữ liệu người nữ trên <Phan_Loai>. 

```python
import pandas as pd
import matplotlib.pyplot as plt

# Tạo DataFrame  
data = {
    "Name": ["Mr. Harris", "Mrs. Bradley", "Miss. Laina", "Mrs. Heath", "Mr. William",
             "Mr. James", "Mr. Timothy", "Mrs. Oscar", "Mrs. Nicholas", "Miss. Marguerite",
             "Mr. William", "Miss. Hulda", "Mr. Lawrence", "Mr. Wilam"],
    "Height": [1.80, 1.61, 1.78, 1.74, 1.55, 1.9, 1.61, 1.40, 1.92, 1.77, 1.55, 1.85, 1.64, 1.77],
    "Weight": [90, 87, 64, 70, 100, 106, 82, 55, 77, 60, 100, 109, 80, 66],
    "Age": [32, 45, 22, 52, 48, 21, 28, 69, 58, 21, 48, 33, 25, 70]
}

df = pd.DataFrame(data)

# Tạo cột GT (Giới tính)
def get_gender(name):
    if name.startswith("Mr."):
        return "Male"
    elif name.startswith("Mrs.") or name.startswith("Miss."):
        return "Female"
    return "Unknown"

df["GT"] = df["Name"].apply(get_gender)

# Xóa tiền tố Mr, Mrs, Miss trong cột Name
df["Name"] = df["Name"].str.replace(r'^(Mr\.|Mrs\.|Miss\.)\s+', '', regex=True)

print(df)

# Lọc dữ liệu người nữ
female_df = df[df["GT"] == "Female"]

# Vẽ biểu đồ cột
plt.figure(figsize=(10, 5))
plt.bar(female_df["Name"], female_df["Height"], color='pink', label='Chiều cao')
plt.bar(female_df["Name"], female_df["Weight"], color='purple', alpha=0.6, label='Cân nặng')

plt.xlabel("Tên")
plt.ylabel("Giá trị")
plt.title("Biểu đồ chiều cao và cân nặng của người nữ")
plt.xticks(rotation=45)
plt.legend()
plt.show()


```
